## Creating Features

```python
def create_features(df):
    '''create time series features from the index '''
    
    df['hour'] = df.index.hour
    df['day_of_week'] = df.index.dayofweek
    df['quarter'] = df.index.quarter
    df['month'] = df.index.month
    df['day_of_month'] = df.index.day
    df['day_of_year'] = df.index.dayofyear
    df['week_of_year'] = df.index.isocalendar().week
    
    return df
```

Applying the function to the dataframe

```python
create_features(df)
```

```python
sns.set()
fig, ax = plt.subplots(figsize=(10, 6))
sns.boxplot(x = 'hour',
           y = 'SystemProduction',
           data = df)
plt.ylabel('Power produced (MW)')
plt.title('Power production by the hour')
plt.show()
```

From the chart, we see that power production starts around 8:00 am and peaks around 14:00 hours, before dropping off again at 21:00 hours

```python
sns.set()
fig, ax = plt.subplots(figsize=(10, 6))
sns.boxplot(x = 'month',
           y = 'SystemProduction',
           data = df)
plt.ylabel('Power produced (MW)')
plt.title('Power production by month')
plt.show()
```

Production of solar energy is lowest in November and December. This makes sense as that is winter season, so the sun isn't out a lot during that time. Power production is very high during the summer i.e between June and August

## Creating the model

```python
train = df.loc[df.index < '2017-10-01']
test = df.loc[df.index >= '2017-10-01']
```

```python
# training set
X_train = train.drop(columns = ['SystemProduction'])
y_train = train['SystemProduction']

# test dataframe
X_test = test.drop(columns = ['SystemProduction'])
y_test = test['SystemProduction']
```

```python
X_train.dtypes
```

Changing the **week_of_year** column in the train and test set to float type, so that the model would have no problems parsing it.

```python
X_train['week_of_year'] = X_train['week_of_year'].astype(np.float64)
X_test['week_of_year'] = X_test['week_of_year'].astype(np.float64)
```

We are going to use XGBoost for predictions.

```python
parameters = {'n_estimators': [5, 10, 100, 1000],
             'max_depth': [2, 4, 6, 8],
             'learning_rate': [0.01, 0.1, 1],
              'max_leaves': [2, 4, 6, 8]
             }
```

```python
model = XGBRegressor()
clf = GridSearchCV(model, parameters, n_jobs = 5, scoring = 'neg_mean_squared_error')
```

```python
clf.fit(X_train, y_train) #fit the model t the train data
```

```python
print('best parameters: ', clf.best_params_)
print('best score: ', clf.best_score_)
```

```python
parameters = {'learning_rate': [0.01], 
              'max_depth': [2], 
              'max_leaves': [2], 
              'n_estimators': [1000]}
model = XGBRegressor()
clf = GridSearchCV(model, parameters, n_jobs = 5, scoring = 'neg_mean_squared_error')
clf.fit(X_train, y_train)
```

```python
# Predictions
y_preds = clf.predict(X_test)
```

```python
print(mean_absolute_error(y_test, y_preds))
```

## Plot feature Importance

```python
reg = XGBRegressor(learning_rate= 0.01, 
              max_depth= 2, 
              max_leaves = 2, 
              n_estimators = 1000)
reg.fit(X_train, y_train)
print(reg.feature_importances_)
```

```python
plot_importance(reg)
plt.show()
```

## Visualizing predictions of one week with test data

```python
# making a copy of test set
test_copy = test.copy()


# adding predictions to the dataframe
test_copy['predictions'] = y_preds


# slicing a week of the data set
fw = test_copy.loc[(test_copy.index <= '2017-10-07')]


# Plotting

sns.set()
plt.figure(figsize=(12, 6))
sns.lineplot(data = fw, x = fw.index, y='SystemProduction', color = 'blue') # actual values
sns.lineplot(data = fw, x = fw.index, y = 'predictions', color = 'black', linestyle = 'dashed') # predicted values
plt.ylabel('Power Produced (MW)')
plt.xlabel('Time(year-month-day)')
plt.show()
```

From the plot, it can be obsrved that the model did not perform so great because power production should be 0 at the ebeginning of a new day.
